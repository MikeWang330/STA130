```python
np.random.seed(130)
reps = 1000
bootstrap_means = []
ffnn_dataset = df_preds[df_preds.hdr_hdicode == 'Low']['ffnn_probability_prediction_error']
xgboost_dataset = df_preds[df_preds.hdr_hdicode == 'Low']['xgboost_probability_prediction_error']
sample_difference = []
for i in range(reps):
    sample1 = ffnn_dataset.sample(150, replace = True)
    sample2 = xgboost_dataset.sample(150, replace = True)
    sample_difference.append((sample1.mean() - sample2.mean()))
confidence_interval = np.quantile(sample_difference,[0.05, 0.95])
print(confidence_interval)
#since the answer include 0, we can conclude xgboost probability prediction error is same with ffnn probability 
#prediction error when hdr hdicode is in 'low' group
```
